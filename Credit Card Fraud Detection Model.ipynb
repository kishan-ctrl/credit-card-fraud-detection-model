```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load dataset
df = pd.read_csv("creditcard.csv")


X = df.drop("Class", axis=1)


constant_cols = X.columns[X.nunique() == 1]
if not constant_cols.empty:
    print(f"Warning: Dropping constant columns: {list(constant_cols)}")
    X = X.drop(columns=constant_cols)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


if not np.all(np.isfinite(X_scaled)):
    print("Warning: Non-finite values detected in X_scaled after standardization. Removing rows containing them.")
    finite_rows_mask = np.all(np.isfinite(X_scaled), axis=1)
    X_scaled = X_scaled[finite_rows_mask]
    if X_scaled.shape[0] == 0:
        raise ValueError("All rows contained non-finite values after scaling. Cannot proceed with PCA.")

# Apply PCA
pca = PCA()
pca.fit(X_scaled)

# Get eigenvalues (explained variance)
eigenvalues = pca.explained_variance_

# Plot Scree Plot
plt.figure(figsize=(8,5))
plt.plot(range(1, len(eigenvalues)+1), eigenvalues, marker='o')

plt.title("Scree Plot")
plt.xlabel("Principal Component")
plt.ylabel("Eigenvalue (Explained Variance)")

# Kaiser criterion line
plt.axhline(y=1, color='red', linestyle='--')

plt.grid()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("creditcard.csv")

# Select some meaningful features
features = ['Amount','V10','V12','V14']

# Calculate mean values by class
class_stats = df.groupby('Class')[features].mean()

# Create grouped bar chart
plt.figure(figsize=(12,6))

class_stats.plot(kind='bar')

plt.title("Class-wise Comparison of Transaction Features")
plt.xlabel("Transaction Class")
plt.ylabel("Average Value")

plt.xticks([0,1], ["Normal (0)", "Fraud (1)"], rotation=0)

plt.legend(title="Features")

plt.tight_layout()
plt.show()
```
